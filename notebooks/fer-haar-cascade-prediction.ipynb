```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import img_to_array

# Load the pre-trained emotion detection model
emotion_model_path = r"C:\Users\Sunil varma\Downloads\FER.h5"
emotion_classifier = load_model(emotion_model_path, compile=False)
EMOTIONS = ['Angry','Disgust','Fear','Happy','Neutral','Sad','Surprise']

# Initialize the face detection cascade classifier
face_detection = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Start video streaming
cv2.namedWindow('Your_Face')
camera = cv2.VideoCapture(0)

while True:
    ret, frame = camera.read()
    if not ret:
        break

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_detection.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30), flags=cv2.CASCADE_SCALE_IMAGE)

    # Process each detected face
    for (fX, fY, fW, fH) in faces:
        # Extract the ROI of the face and preprocess it for classification
        roi_gray = gray[fY:fY + fH, fX:fX + fW]
        roi_gray = cv2.resize(roi_gray, (100, 100))
        roi_gray = roi_gray.astype("float") / 255.0
        roi_gray = img_to_array(roi_gray)
        roi_gray = np.expand_dims(roi_gray, axis=0)

        # Perform emotion classification on the ROI
        preds = emotion_classifier.predict(roi_gray)[0]
        emotion_label = EMOTIONS[np.argmax(preds)]
        emotion_prob = np.max(preds) * 100

        # Draw a rectangle around the detected face and display the emotion label
        cv2.rectangle(frame, (fX, fY), (fX + fW, fY + fH), (0, 255, 0), 2)
        cv2.putText(frame, f"{emotion_label}: {emotion_prob:.2f}%", (fX, fY - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.45, (0, 255, 0), 2)

    # Display the frame
    cv2.imshow('Your_Face', frame)

    # Check for the 'q' key press to exit the loop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close all windows
camera.release()
cv2.destroyAllWindows()
```
